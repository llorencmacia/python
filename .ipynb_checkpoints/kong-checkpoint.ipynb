```python
def chocked_mass_flow (pressure, temp,  area):
        gamma = 1.4
        if temp-zero_Celsius < 0:
            print("has de sumar-hi el conversor de k a ºC")
        else:    
            air = Mixture( 'air' , T = temp , P = pressure )
            r   = sci.constants.R / air.MW * 1000
   
            d   = gamma+1
            e   = 2*(gamma-1)
    
            A   = pressure * area
            B   = (r*temp) **(-0.5)
            C   = (gamma ) **0.5
            D   = (  2 / d  )**(d/e)
            
        return  A * B * C * D

def area(x):
    return np.pi*x**2/4

def perimetre(x):
    return np.pi*x

def p_0_p_est (Mach):
    gamma = 1.4
    a = gamma /(gamma-1)
    b = (gamma -1)/2
    return (1 + b * Mach**2)**a
def t_0_t_est (Mach):
    gamma = 1.4
    b = (gamma -1)/2
    return (1 + b * Mach**2)

def get_A_by_Astar_from_M(M):
#         """Returns the value of A/A* at Mach number M"""
    g = 1.4
    if M < 0:
        print ("Value of M negative. M value should be positive")
        return
    else:
        return 1.0/M*(2/(g+1)*(1+(g-1)*0.5*M**2))**((g+1)/(g-1)*0.5)

def get_M_from_A_by_Astar(A_by_Astar):
    return (optimize.bisect(lambda M: get_A_by_Astar_from_M(M)-A_by_Astar, 0.001, 1.0),optimize.bisect(lambda M: get_A_by_Astar_from_M(M) - A_by_Astar, 1.0, 100))
```

```python
P_o_1 = 7e5
T_o_1 = 300
d_ch_t = 2.75e-3
d_ch_f = 7e-3

a_ch_t = area(d_ch_t)
a_ch_f = area(d_ch_f)
## 1r bloc
m_dot_1 = chocked_mass_flow(P_o_1,T_o_1, a_ch_t)

#2nd Bloc
M_ch_1 = 1
A_by_Astar = a_ch_f/a_ch_t

p_ch_t = P_o_1 / p_0_p_est(M_ch_1)
t_ch_t = T_o_1 / t_0_t_est(M_ch_1)
mach_ch_f = get_M_from_A_by_Astar(A_by_Astar)


```
