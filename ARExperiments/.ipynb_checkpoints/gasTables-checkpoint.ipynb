```python
import numpy
from   thermo             import Mixture
from   scipy.constants    import atmosphere, zero_Celsius, mmHg, g
import scipy           
import numpy
from   thermo             import Mixture
from   scipy.constants    import atmosphere, zero_Celsius, mmHg, g
import scipy           
from   scipy              import optimize
from   math               import *
import math

def area ( diam ):
    return  ( math.pi * diam ** 2 ) / 4
def diametre ( area ):
    return  ( area * 4 / math.pi ) ** ( 1/2 )

class Isentropic:
    def __init__(self, g=1.4):
        # g is the value of gamma (ratio of specific heats), default = 1.4
        self.g = g
        
    def get_P_by_Po_from_M(self, M):
       # Returns the value of p/po at Mach number M"""
        if M < 0:
            print ('Value of M negative. M value should be positive')
            return
        else:
            return (1 + (self.g -1 )*0.5*M**2)**(self.g/(1-self.g))
    
    def get_AP_by_AstarPo_from_M(self, M):
#         """Returns the value of Ap/A*po at Mach number M"""
        if M < 0:
            print ('Value of M negative. M value should be positive')
            return
        else:
            return ((2/(self.g+1))**((self.g+1)/(self.g-1)*0.5))/(M * sqrt(1 + (self.g-1)*0.5 * M**2))
    
    def get_A_by_Astar_from_M(self, M):
#         """Returns the value of A/A* at Mach number M"""
        if M < 0:
            print ('Value of M negative. M value should be positive')
            return
        else:
            return 1.0/M*(2/(self.g+1)*(1+(self.g-1)*0.5*M**2))**((self.g+1)/(self.g-1)*0.5)

    def get_M_from_A_by_Astar(self, A_by_Astar):
        #"""Returns the two values of M possible(subsonic and supersonic) at A/A* equal to A_by_Astar"""
        if A_by_Astar < 1.0:
            print ("A/A* must be greater than 1")
            return
        elif A_by_Astar == 1:
            return 1.0
        else:
            return (optimize.bisect(lambda M: self.get_A_by_Astar_from_M(M)-A_by_Astar, 0.001, 1.0),optimize.bisect(lambda M: self.get_A_by_Astar_from_M(M) - A_by_Astar, 1.0, 100))

    def get_Mstar_from_M(self, M):
#         """Returns the value of M* at Mach number M"""
        if M < 0:
            print ('Value of M negative. M value should be positive')
            return
        else:
            return sqrt(((self.g+1)*0.5*M**2)/(1 + (self.g-1)*0.5*M**2))
    def get_M_from_P_by_Po(self, P_by_Po):
#         """Returns the value of M at p/po equal to P_by_Po"""
        if P_by_Po <= 0.0 or P_by_Po >= 1.0:
            print ("p/p0 must be between 0 and 1")
            return
        else:
            return sqrt(2*((1/((P_by_Po)**((self.g-1)/self.g)))-1)/(self.g-1))
    
    def chocked_mass_flow (self, pressure, temp,  area):
        if temp-zero_Celsius < 0:
            print("has de sumar-hi el conversor de k a ÂºC")
        else:    
            air = Mixture( 'air' , T = temp , P = pressure )
            r   = scipy.constants.R / air.MW * 1000
   
            d   = self.g+1
            e   = -2*(d-2)
    
            A   = pressure * area
            B   = (  1+ (d-2) /2  )**(d/e)
            C   = (self.g / (r* temp)) **0.5
    
        return  A * B * C 
    def pressure_from_mach_p0 ( M , p0 , gamma ):
        x = Symbol('x')
    
        a = gamma - 1
        b = gamma / a
    
        pressure_from_mach_p0 = solve( - p0 + x * (1 + a * M*M /2 ) ** b , x )
        just = (1 + a * M*M /2 ) ** - b
        return pressure_from_mach_p0, just

    def area_relation_from_mach ( self, M1 , M2 ):
        
        a = ( self.g - 1 ) / 2
        b = a * M1 ** 2
        c = a * M2 ** 2
        d = ( 1 + c ) / ( 1 + b)
        e = d ** ( (self.g + 1) / (4* a) ) 
    
        return  M1 * e / M2
    def area_from_machstar (  M ):
        return area_relation_from_mach ( 1 , M ) 
    def mach_from_p0_p ( p0 , p , gamma):
        x = Symbol('x')
    
        a = gamma - 1
        b = gamma / a
    
        return   ( (((p0 /p)**(a/gamma)) - 1 ) * 2 / a )**0.5  
    
class NormalShock:   
    def __init__(self, g=1.4):
#         """ g is the value of gamma (ratio of specific heats), default = 1.4
#         """
            self.g = g
            self.Isentropic = Isentropic(self.g)
        
    def get_My_from_Mx(self, Mx):
#         """Returns Mach No My behind a normal shock from Mach No Mx in front of the shock"""
        if Mx <= 1:
            print ("Mach No ahead of ths shock Mx should be greater than 1")
        else:
            return sqrt((2/(self.g-1) + Mx**2)/(2*self.g/(self.g-1)*Mx**2 -1))
    
    def get_Py_by_Px_from_Mx(self, Mx):
#         """Returns py/px across a normal shock from  Mx """
        if Mx <= 1:
            print ("Mach No ahead of ths shock Mx should be greater than 1")
        else:
            return 2*self.g/(self.g+1)* Mx**2 - (self.g-1)/(self.g+1)
            
    def get_Poy_by_Pox_from_Mx(self, Mx):
#         """Returns poy/pox across a normal shock from Mx"""
        if Mx <= 1:
            print ("Mach No ahead of ths shock Mx should be greater than 1")
        else:
            My = self.get_My_from_Mx(Mx)
            return self.Isentropic.get_P_by_Po_from_M(Mx) * self.get_Py_by_Px_from_Mx(Mx) / self.Isentropic.get_P_by_Po_from_M(My)        
            
```
